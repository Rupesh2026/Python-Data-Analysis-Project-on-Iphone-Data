```python
df['Model Name']=df['Product Name'].str[6:15] # Adding New column 
df.head()
```

```python
df.sort_values(by=['Star Rating'], ascending=False).head(10) # Sorting by best rated Iphone
```

```python
df.groupby('Model Name')['Discount Percentage'].mean().reset_index()  # Average discount percentage by Model Name
```

```python
df.groupby('Model Name')['Star Rating'].mean().reset_index().sort_values('Star Rating', ascending=False).head(5) #Highest average star rating
```

```python
df['Price Diff'] = df['Mrp'] - df['Sale Price'] 
df.sort_values('Price Diff', ascending=False).head(5)
# products with highest difference between MRP and Sale Price
```
